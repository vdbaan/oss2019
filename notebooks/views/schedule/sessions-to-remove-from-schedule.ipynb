```python
#%load_ext autoreload
#%autoreload

pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows'    , 1000)

from oss_hugo.OSS_Schedule import OSS_Schedule
schedule    = OSS_Schedule()
def get_sessions_to_remove():
    df_sessions = schedule.sessions_mapped_by_size()
    query = df_sessions['organizers'].apply(lambda x: len(x) == 0)
    result = df_sessions[query]
    result = result.sort_values('track').reset_index(drop=True)
    result.index += 1
    return result
```
